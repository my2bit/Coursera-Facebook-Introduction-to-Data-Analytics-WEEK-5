Question 8: Best Products
What are the top ten product titles by the number of sales? Display in descending order. Store in variable top_num_sales.

```python
# Number of unique sales a product was ordered
df_cleaned['prod_title'].value_counts()[:3]
top_num_sales = df_cleaned['prod_title'].value_counts()[:5]
top_num_sales
```

```python
# Number of unique sales a product was ordered
top_num_sales = df_cleaned.groupby('prod_title')['trans_quantity'].count().nlargest(5)
top_num_sales
```

```python
# total quantity of a product orderd
top_num_sales = df_cleaned.groupby('prod_title')['trans_quantity'].sum().nlargest(5)
top_num_sales
```

```python
# Filters rows that include 'Reddy Beddy'
df_cleaned['prod_title'] == 'Reddy Beddy'
```

```python
# Applying the filter to get a datframe consisting only of Reddy Beddy

df_reddy_beddy = df_cleaned[df_cleaned['prod_title'] == 'Reddy Beddy']
df_reddy_beddy
```

```python
# grouping by number of units of Reddy Beddy (1 - 6) included in an order
df_reddy_beddy.groupby('trans_quantity')['prod_title'].count()
```
